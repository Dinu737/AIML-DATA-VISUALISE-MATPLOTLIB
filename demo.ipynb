```python
from sklearn.linear_model import LinearRegression

import pandas as pd

data = {
    "model_year":[2020, 2021, 2022, 2023],
    "engine_size":[1.2,1.3,1.1,1.4],
    "milage":[100,200,300,201],
    "cost":[20000,1000,3000,2332] 
}

df = pd.DataFrame(data)
print(df)

X = df[["model_year","engine_size","milage"]]
y = df[["cost"]]

model = LinearRegression()
model.fit(X,y)

new_car = pd.DataFrame({
    "model_year": [2024],
    "engine_size": [1.5],
    "milage": [2244]
})

predicted_cost = model.predict(new_car)

predicted_cost = model.predict(new_car)
print("Predicted Cost: ₹", round(predicted_cost[0][0], 2))
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Sample Data
data = {
    "model_year": [2022, 2021, 2023, 2020, 2024],
    "engine_size": [1.2, 1.2, 1.5, 1.0, 1.5],
    "mileage": [20000, 30000, 10000, 40000, 5000],
    "cost": [650000, 700000, 900000, 500000, 1000000]
}

df = pd.DataFrame(data)

# Independent Variables (X)
X = df[["model_year", "engine_size", "mileage"]]

# Dependent Variable (Y)
y = df["cost"]

# Create and Train Model
model = LinearRegression()
model.fit(X, y)

# New Car Details
new_car = pd.DataFrame({
    "model_year":[2024],
    "engine_size":[2.4],
    "mileage":[23233]
})

# Predict Cost
predicted_cost = model.predict(new_car)

print("Predicted Cost: ₹", round(predicted_cost[0], 2))
```
